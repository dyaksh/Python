```python
## 228. Compute mathematical operations on Array, Add & Multiply two matrices
import numpy as np

# Function to take user input for a matrix
def input_matrix(rows, columns):
    print(f"Enter the elements of the {rows}x{columns} matrix row-wise:")
    matrix = []
    for i in range(rows):
        row = list(map(int, input().split()))
        matrix.append(row)
    return np.array(matrix)

# Function to perform addition and multiplication operations on two matrices
def matrix_operations(matrix1, matrix2):
    # Addition
    addition_result = matrix1 + matrix2
    # Multiplication
    multiplication_result = np.dot(matrix1, matrix2)
    return addition_result, multiplication_result

# Take user input for the dimensions of the matrices
rows1 = int(input("Enter the number of rows in the first matrix: "))
columns1 = int(input("Enter the number of columns in the first matrix: "))
rows2 = int(input("Enter the number of rows in the second matrix: "))
columns2 = int(input("Enter the number of columns in the second matrix: "))

# Validate matrix dimensions for multiplication
if columns1 != rows2:
    print("Matrices cannot be multiplied. The number of columns in the first matrix must be equal to the number of rows in the second matrix.")
else:
    # Input first matrix
    print("\nEnter the elements of the first matrix:")
    matrix1 = input_matrix(rows1, columns1)
    
    # Input second matrix
    print("\nEnter the elements of the second matrix:")
    matrix2 = input_matrix(rows2, columns2)
    
    # Perform addition and multiplication operations
    addition_result, multiplication_result = matrix_operations(matrix1, matrix2)
    
    # Print results
    print("\nAddition Result:")
    print(addition_result)
    print("\nMultiplication Result:")
    print(multiplication_result)
```

```python
#29. Find the most frequent value in a NumPy array  
import numpy as np

# Function to find the most frequent value in a NumPy array
def most_frequent_value(array):
    # Flatten the array to 1D
    flattened_array = array.flatten()
    # Count occurrences of each value
    counts = np.bincount(flattened_array)
    # Find the value with the highest count
    most_frequent = np.argmax(counts)
    return most_frequent

# Take user input for the dimensions and elements of the array
rows = int(input("Enter the number of rows in the array: "))
columns = int(input("Enter the number of columns in the array: "))

print(f"Enter the elements of the {rows}x{columns} array:")

# Take input for the elements of the array
array_rows = []
for i in range(rows):
    row = list(map(int, input().split()))
    array_rows.append(row)

# Convert the list of rows into a NumPy array
array = np.array(array_rows)

# Find the most frequent value in the array
result = most_frequent_value(array)

# Print the most frequent value
print("Most frequent value in the array:", result)
```

```python
#35  WAP to Reverse a Number
def reverse_number(number):
    reversed_number = 0
    while number > 0:
        # Get the last digit of the number
        digit = number % 10
        # Add the digit to the reversed number (multiply by 10 to shift the digits to the left)
        reversed_number = (reversed_number * 10) + digit
        # Remove the last digit from the number
        number = number // 10
    return reversed_number

# Take input from the user
number = int(input("Enter a number: "))

# Reverse the number
reversed_number = reverse_number(number)

# Print the reversed number
print("Reversed number:", reversed_number)
```

```python
#37.WAP to determine all Pythagorean triplets
# Take input from the user for the upper limit
limit = int(input("Enter the upper limit for finding Pythagorean triplets: "))

# Find and print all Pythagorean triplets up to the limit
triplets = []
for a in range(1, limit):
    for b in range(a, limit):
        c = (a**2 + b**2)**0.5
        if c.is_integer():
            triplets.append((a, b, int(c)))

print("Pythagorean Triplets up to limit", limit, ":")
for triplet in triplets:
    print(triplet)
```

```python
'''39.Given a list A of N distinct integer numbers, you can sort the list by moving an element to the end of the list. Find the minimum number
of moves required to sort the list using this method in ascending order'''
def min_moves_to_sort(A):
    # Find the length of the longest increasing subsequence (LIS)
    LIS = [1] * len(A)
    for i in range(1, len(A)):
        for j in range(0, i):
            if A[i] > A[j] and LIS[i] < LIS[j] + 1:
                LIS[i] = LIS[j] + 1
    max_length_LIS = max(LIS)
    
    # Minimum number of moves required is equal to the length of the list minus the length of the LIS
    return len(A) - max_length_LIS

# Example usage:
A = [5, 2, 6, 3, 4, 1]
print("Minimum number of moves required:", min_moves_to_sort(A))


# Take input from the user for the list of integers
A = list(map(int, input("Enter the list of integers separated by space: ").split()))

# Find the length of the longest increasing subsequence (LIS)
LIS = [1] * len(A)
for i in range(1, len(A)):
    for j in range(0, i):
        if A[i] > A[j] and LIS[i] < LIS[j] + 1:
            LIS[i] = LIS[j] + 1
max_length_LIS = max(LIS)

# Minimum number of moves required is equal to the length of the list minus the length of the LIS
min_moves = len(A) - max_length_LIS

# Print the minimum number of moves required
print("Minimum number of moves required to sort the list:", min_moves)
```

```python
#63.Write a Python class to reverse a string word by word
class WordReverser:
    def __init__(self, text):
        """
        Initialize the WordReverser with a specific text.

        Args:
        text (str): The string to be reversed word by word.
        """
        self.text = text

    def reverse_words(self):
        """
        Reverse the order of words in the string provided at initialization.

        Returns:
        str: A new string with the words in reverse order.
        """
        # Split the string into words
        words = self.text.split()

        # Reverse the list of words
        reversed_words = words[::-1]

        # Join the reversed list of words back into a string
        reversed_text = ' '.join(reversed_words)

        return reversed_text

    def __str__(self):
        """
        String representation of the initial text.

        Returns:
        str: The initial text.
        """
        return self.text

# Example usage:
original_text = "hello world"
reverser = WordReverser(original_text)
reversed_text = reverser.reverse_words()

print("Original text:", original_text)
print("Reversed text:", reversed_text)
```

```python
''' 68.WAPP to push all zeros to the end of a given list. The order of elements should not change:
 Input: 0 2 3 4 6 7 9 0
 Output:2 3 4 6 7 9 0 0'''

def push_zeros_to_end(lst):
    """
    This function takes a list of numbers and moves all zeros to the end
    while preserving the order of non-zero elements.

    Args:
    lst (list): The list of numbers.

    Returns:
    list: Modified list with all zeros pushed to the end.
    """
    non_zero = [x for x in lst if x != 0]  # List to store non-zero elements
    zero_count = lst.count(0)  # Count the zeros in the list

    # Return the list of non-zero elements followed by zeros
    return non_zero + [0] * zero_count

# Example usage:
input_list = [0, 2, 3, 4, 6, 7, 9, 0]
result_list = push_zeros_to_end(input_list)
print("Output:", ' '.join(map(str, result_list)))
```

```python
'''69 Write a program that accepts a comma separated sequence of words as input and prints the words in a comma separated sequence after
sorting them alphabetically.
 Input: without, hello,bag, world
 Output: bag,hello,without,world'''
def sort_words(input_string):
    """
    Sorts the words in a comma-separated string alphabetically.

    Args:
    input_string (str): A comma-separated string of words.

    Returns:
    str: A comma-separated string of words sorted alphabetically.
    """
    # Split the input string into a list of words
    words = input_string.split(',')

    # Strip whitespace and sort the list of words
    sorted_words = sorted(word.strip() for word in words)

    # Join the sorted list back into a comma-separated string
    return ','.join(sorted_words)

# Get input from the user
input_string = input("Enter a comma separated sequence of words: ")

# Call the function and print the sorted words
sorted_string = sort_words(input_string)
print("Sorted words:", sorted_string)
```

```python
''' Q. 71 With given list L of integers , write a program that prints this L after removing duplicate values with original order preserved. 
 Input: 12 24 35 24 88 120 155 88 120 155 
 Output: 12 24 35 88 120 155
 '''
def remove_duplicates(L):
    seen = set()
    result = []
    for item in L:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

def main():
    input_str = input("Enter the list of integers separated by space: ")
    L = list(map(int, input_str.split()))
    result = remove_duplicates(L)
    print("Output:", ' '.join(map(str, result)))

if __name__ == "__main__":
    main()
```

```python
''' Q. 74 Python program to generate the prime numbers from 1 to N'''
def generate_primes(N):
    primes = []
    for num in range(2, N-1):
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
    return primes

def main():
    N = int(input("Enter the value of N: "))
    prime_numbers = generate_primes(N)
    print("Prime numbers from 1 to", N, "are:", prime_numbers)

if __name__ == "__main__":
    main()
```

```python
''' Q.77 Write a Python class to convert a roman numeral to an integer'''
class RomanToInt:
    def __init__(self):
        self.roman_map = {
            'I': 1,
            'V': 5,
            'X': 10,
            'L': 50,
            'C': 100,
            'D': 500,
            'M': 1000
        }
    
    def roman_to_int(self, s: str) -> int:
        result = 0
        prev_value = 0
        for char in s[::-1]:
            value = self.roman_map[char]
            if value < prev_value:
                result -= value
            else:
                result += value
            prev_value = value
        return result

# Example usage:
def main():
    converter = RomanToInt()
    roman_numeral = input("Enter a Roman numeral: ")
    result = converter.roman_to_int(roman_numeral)
    print("Integer value:", result)

if __name__ == "__main__":
    main()
```

```python
''' Q. 78 Write a Python class to get all possible unique subsets from a set of distinct integers'''
class SubsetGenerator:
    def __init__(self):
        self.subsets = []
    
    def generate_subsets(self, nums):
        self.subsets = []
        self._generate(nums, [], 0)
        return self.subsets
    
    def _generate(self, nums, subset, index):
        self.subsets.append(subset[:])
        for i in range(index, len(nums)):
            subset.append(nums[i])
            self._generate(nums, subset, i + 1)
            subset.pop()

# Example usage:
def main():
    generator = SubsetGenerator()
    nums_str = input("Enter numbers separated by spaces: ")
    nums = list(map(int, nums_str.split()))
    result = generator.generate_subsets(nums)
    print("All possible unique subsets of", nums, "are:")
    for subset in result:
        print(subset)

if __name__ == "__main__":
    main()
```

```python
'''Q. 79 write a Python class to find a pair of elements (indices of the two numbers) from a given array whose sum equals a specific target 
number'''
class TwoSum:
    def __init__(self):
        pass
    
    def find_indices(self, nums, target):
        num_map = {}
        for i, num in enumerate(nums):
            complement = target - num
            if complement in num_map:
                return [num_map[complement], i]
            num_map[num] = i
        return None

# Example usage:
def main():
    finder = TwoSum()
    nums_str = input("Enter numbers separated by spaces: ")
    nums = list(map(int, nums_str.split()))
    target = int(input("Enter the target number: "))
    indices = finder.find_indices(nums, target)
    if indices:
        print("Indices of the two numbers:", indices)
        print("Numbers:", nums[indices[0]], nums[indices[1]])
    else:
        print("No pair found.")

if __name__ == "__main__":
    main()
```

```python
''' Q. 80 write a Python class to implement pow(x, n).'''
class PowerCalculator:
    def __init__(self):
        pass
    
    def power(self, x, n):
        if n == 0:
            return 1
        elif n < 0:
            return 1 / self.power(x, -n)
        else:
            if n % 2 == 0:
                half_pow = self.power(x, n // 2)
                return half_pow * half_pow
            else:
                half_pow = self.power(x, (n - 1) // 2)
                return x * half_pow * half_pow

# Example usage:
def main():
    calculator = PowerCalculator()
    x = float(input("Enter the base number (x): "))
    n = int(input("Enter the power (n): "))
    result = calculator.power(x, n)
    print("Result of", x, "raised to the power of", n, "is:", result)

if __name__ == "__main__":
    main()
```

```python
''' Q.81 Write a NumPy program to convert the values of Centigrade degrees into Fahrenheit degrees and vice versa. Values are stored into a 
NumPy array. Sample Array [0, 12, 45.21, 34, 99.91]'''
import numpy as np

class TemperatureConverter:
    def __init__(self):
        pass
    
    def celsius_to_fahrenheit(self, celsius_array):
        return celsius_array * 9 / 5 + 32
    
    def fahrenheit_to_celsius(self, fahrenheit_array):
        return (fahrenheit_array - 32) * 5 / 9

# Example usage:
def main():
    converter = TemperatureConverter()
    
    # Take input from the user
    nums_str = input("Enter temperatures in Celsius separated by spaces: ")
    celsius_array = np.array(list(map(float, nums_str.split())))
    
    # Convert Celsius to Fahrenheit
    fahrenheit_array = converter.celsius_to_fahrenheit(celsius_array)
    print("Celsius to Fahrenheit:", fahrenheit_array)
    
    # Take input from the user
    nums_str = input("Enter temperatures in Fahrenheit separated by spaces: ")
    fahrenheit_array = np.array(list(map(float, nums_str.split())))
    
    # Convert Fahrenheit to Celsius
    celsius_array = converter.fahrenheit_to_celsius(fahrenheit_array)
    print("Fahrenheit to Celsius:", celsius_array)

if __name__ == "__main__":
    main()
```

```python
''' Q. 82 write a NumPy program to find the set difference of two arrays. The set difference will return the sorted, unique values in array1 that are 
not in array2'''
import numpy as np

class SetDifferenceFinder:
    def __init__(self):
        pass
    
    def find_set_difference(self, array1, array2):
        return np.setdiff1d(array1, array2)

# Example usage:
def main():
    finder = SetDifferenceFinder()
    
    # Take input from the user for array1
    array1_str = input("Enter elements of array1 separated by spaces: ")
    array1 = np.array(list(map(int, array1_str.split())))
    
    # Take input from the user for array2
    array2_str = input("Enter elements of array2 separated by spaces: ")
    array2 = np.array(list(map(int, array2_str.split())))
    
    # Find set difference
    result = finder.find_set_difference(array1, array2)
    print("Set difference:", result)

if __name__ == "__main__":
    main()
```
